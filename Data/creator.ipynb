```python
def generate_data_set(num_days = 20, num_sites= 1):

    # Function to generate a range of dates
    def generate_dates(start_date, num_days):
        for n in range(num_days):
            yield start_date + timedelta(n)

    # Parameters
    sites = [x+1 for x in range(num_sites)]
    room_types = ['Single','Double','Premium']
    start_date = datetime(2024, 6, 24)


    # Generate data
    data = []
    
    single_price = np.random.randint(100, 150)
    double_price = np.random.randint(200, 300)
    suite_price = np.random.randint(500, 1200)
        
    for site in sites:
        for date in generate_dates(start_date, num_days):
            data.append([site, room_types[0], date.strftime('%Y-%m-%d'), single_price])
            data.append([site, room_types[1], date.strftime('%Y-%m-%d'), double_price])
            data.append([site, room_types[2], date.strftime('%Y-%m-%d'), suite_price])

    # Create DataFrame
    df = pd.DataFrame(data, columns=['site', 'type', 'date', 'price'])
    plt.figure(figsize=(12, 6)) 
    sns.lineplot(data = df, x = 'date', y='price',hue = 'site')
    plt.xticks(rotation=45)
    plt.show()
    return df.pivot(index=['type','site'],columns=['date'],values=['price'])
    

df = generate_data_set(20,3)
df
```
